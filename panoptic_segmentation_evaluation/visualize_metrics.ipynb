```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Load metrics
metrics_df = pd.read_csv("/home/giuliano/mt_ipp_panoptic_mapping/datasets/scannetv2-dvc/scans/scene0144_01/max_deeplab_l_panoptic_labels_metrics.csv")
metrics_df.head(10)
```

```python

fig, axs = plt.subplots(2, 3, figsize=(25,15))
for i, metric_name in enumerate(metrics_df.columns.to_list()):
    r = i // 3
    c = i % 3
    ax = axs[r, c]
    ax.title.set_text(metric_name)
    ax.grid()
    data = metrics_df[metric_name]
    sns.lineplot(data=data, ax=ax)    
    
```
